```python
import sympy
import math 
import numpy as np

x = sympy.Symbol('x')
y = sympy.Function('y')
c = sympy.Symbol('c')
b = sympy.Symbol('b')

# expreso la ecuacion
f = (y(x))*(1-y(x))*((2*(y(x)**0.5)/c + 1)) - 3*y(x)*b**2
                     
sympy.Eq(y(x).diff(x), f)

```

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt

c1, c2, c3, c4 = 1, 1.15, 1.4, 1.5 
b1, b2, b3, b4 = 0.001, 0.3, 0.5, 0.8

# function that returns dy/dt
def model1 (y,x):

    dydx = y*(1-y)*((2*(np.sqrt(y)))/c2 + 1) - 3*y*b1**2
                     
    return dydx

def model2 (y,x):

    dydx = y*(1-y)*((2*(np.sqrt(y)))/c2 + 1) - 3*y*b2**2
                     
    return dydx

def model3 (y,x):

    dydx = y*(1-y)*((2*(np.sqrt(y)))/c2 + 1) - 3*y*b3**2
                     
    return dydx
def model4 (y,x):

    dydx = y*(1-y)*((2*(np.sqrt(y)))/c2 + 1) - 3*y*b4**2
                     
    return dydx

# initial condition
y0 = 0.0001

# time points
x = np.linspace(math.log(10**(-6)),math.log(10))

# solve ODE
y1 = odeint(model1,y0,x)
y2 = odeint(model2,y0,x)
y3 = odeint(model3,y0,x)
y4 = odeint(model4,y0,x)
# plot results
plt.plot(x,y1,color='blue',label='b=0.001')
plt.plot(x,y2,color='red',label='b = 0.3')
plt.plot(x,y3,color='green',label='b=0.5')
plt.plot(x,y4,color='violet',label='b=0.8')
plt.xlabel('ln(a)')
plt.legend()
plt.title('Holographic dark energy density parameter with $c=1.15$')
plt.ylabel('$\Omega_{de}$')
plt.grid()
```
