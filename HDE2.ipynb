```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import numpy as np
```

```python
from simplemc.cosmo.Parameter import Parameter
from simplemc.DriverMC import DriverMC
```

$$  

H^2(a) = H_0^2 \left( \Omega_{r,0} a^{-4} + \Omega_{m,0} a^{-3} + \Omega_{k,0} a^{-2} + \Omega_{\Lambda , 0 } a^{-2+2/c}\right) 
$$

```python
cextra = Parameter('cextra', 0.7, 0.001, (0.5,1.5), 'c_{extra}')
Ok = Parameter('Ok', 0.0,0.1,(-0.02,0.02), '\Omega_k') 
paramlist=[cextra,Ok]
cosmo_model = 'Omrad/a**4+Om/a**3+Ok/a**2+(1-Omrad-Om-Ok)*a**(-2+2/cextra)'
analyzer = DriverMC(model='simple_cosmo', datasets='BPantheon+Planck+CBAO', analyzername='nested',
                    custom_parameters=paramlist, custom_function=cosmo_model)
```

```python
res = analyzer.executer(nlivepoints=200)
```

```python
samples = res['result']['samples']
weights = res['result']['weights']
```

```python
labels =  ['{}'.format(p.Ltxname) for p in analyzer.pars_info]
labels, np.shape(samples)
```

```python
samp1 = MCSamples(samples=samples, names=labels, labels = labels, label='h')
samp2 = MCSamples(samples=samples, names = labels, labels = labels, label='cextra')
```

```python
samp =MCSamples(samples=samples, names=labels, labels = labels)
```

```python
g = plots.get_subplot_plotter()
g.triangle_plot([samp], filled=False, line_args={'lw':1},
                colors=['green', 'red', 'indigo'],
                settings={ 'smooth_scale_2D':0.8, 'smooth_scale_1D':0.6},)
```

```python
g1 = plots.get_single_plotter(width_inch=6, ratio= 2. / 3.)
g1.plot_2d([samp2,samp1], labels[2],labels[3], filled= True)
g1.add_legend(['SN','CBAO'], legend_loc='upper right',colored_text=True);
```
